```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('basic_dataset.xls', header=1, index_col='ID')
```

```python
# --- Initial Inspection ---
display(df.head())

print("\n--- Data Types and Non-Null Counts ---")
df.info()

print("\n--- Statistical Summary ---")
display(df.describe())
```

```python
# --- Standardize Column Names ---
df.rename(columns={'default payment next month': 'default_payment'}, inplace=True)
df.columns = [x.lower() for x in df.columns]

# --- Correct Categorical Inconsistencies ---
# Combine undocumented values in 'education' and 'marriage' into the 'others' category
df['education'] = df['education'].replace([0, 5, 6], 4)
df['marriage'] = df['marriage'].replace(0, 3)

# --- Check for Null Values ---
# This confirms there are no missing values to handle
print(f"Total null values: {df.isnull().sum().sum()}")

# --- Handle Duplicates (in the correct order) ---
# This step simplifies payment status, which can create duplicates
pay_cols = ['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']
for col in pay_cols:
    df[col] = df[col].replace([-2, -1], 0)

# Now, find and remove the duplicates that may have been created
print(f"Found {df.duplicated().sum()} duplicate rows.")
df.drop_duplicates(inplace=True)
print(f"Duplicates remaining after removal: {df.duplicated().sum()}")

print("\nInitial data cleaning complete.")
print(f"Shape of DataFrame after cleaning: {df.shape}")
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='default_payment', data=df)
plt.title('Default Payment Distribution (0: No, 1: Yes)')

plt.subplot(1, 2, 2)
sns.countplot(x='education', data=df)
plt.title('Education Level Distribution')
plt.show()

"""What to look for: The first plot shows a class imbalance; far more customers did not default than those who did. 
This is important to know for modeling.
"""
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['limit_bal'], kde=True, bins=30)
plt.title('Credit Limit (limit_bal) Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['limit_bal'])
plt.title('Credit Limit (limit_bal) Box Plot')
plt.show()

"""What to look for: The histogram is right-skewed, meaning most people have lower credit limits. 
The box plot clearly shows many data points above the main whisker; these are our outliers."""
```

```python
# Scatter Plot (Numerical vs. Numerical)
# Let's see if there's a relationship between age and limit_bal

sns.scatterplot(x='age', y='limit_bal', data=df, alpha=0.3)
plt.title('Age vs. Credit Limit')
plt.show()

# What to look for: There's no strong, clear linear relationship, but we can see that credit limits are spread across all age group
```

```python
# Select a feature to analyze
col_to_analyze = 'bill_amt1'

# Calculate and print its original skewness
original_skewness = df[col_to_analyze].skew()
print(f"--- Outlier Analysis for '{col_to_analyze}' ---")
print(f"Original Skewness: {original_skewness:.4f}")

sns.histplot(df[col_to_analyze], kde=True)
plt.title(f'Original Distribution of {col_to_analyze}')
plt.show()

# Observation: The initial skewness is high and positive, confirming a long right tail caused by outliers, which we also saw in the box plots.
```

```python
# Create a copy of the DataFrame for testing
df_test = df.copy()

# --- Identify Outliers using the IQR method ---
Q1 = df_test[col_to_analyze].quantile(0.25)
Q3 = df_test[col_to_analyze].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_mask = (df_test[col_to_analyze] < lower_bound) | (df_test[col_to_analyze] > upper_bound)

# --- Option A: Replace Outliers with the Median ---
median_val = df_test[col_to_analyze].median()
df_test['median_imp'] = df_test[col_to_analyze].where(~outlier_mask, median_val)
median_imp_skew = df_test['median_imp'].skew()

# --- Option B: Capping (Winsorizing) at 1st and 99th Percentiles ---
p1 = df_test[col_to_analyze].quantile(0.01)
p99 = df_test[col_to_analyze].quantile(0.99)
df_test['capped'] = np.clip(df_test[col_to_analyze], p1, p99)
capped_skew = df_test['capped'].skew()

# --- Option C: Log Transformation ---
# We shift the data to ensure all values are positive before applying the log
df_test['log_transform'] = np.log1p(df_test[col_to_analyze] - df_test[col_to_analyze].min())
log_skew = df_test['log_transform'].skew()

# Create a summary DataFrame
results = pd.DataFrame({
    'Method': ['Original Data', 'Replace with Median', 'Capping (Winsorizing)', 'Log Transform'],
    'Skewness': [original_skewness, median_imp_skew, capped_skew, log_skew]
}).sort_values(by='Skewness', key=abs)

print("--- Skewness Comparison ---")
print(results)

"""Analysis and Decision:

Replacing with Median and Log Transform both created new problems (heavy distortion or overcorrection to a negative skew), as we saw in our previous detailed analysis.

Capping (Winsorizing) is the clear winner. It dramatically reduced skewness to 0.07, which is very close to symmetrical, without artificially distorting the data's underlying distribution.

Conclusion: The best and most robust option is Capping."""
```

```python
print("Applying the best option (Capping) to the original DataFrame...")

# Define all numerical columns that need capping
cols_to_cap = ['bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6',
               'pay_amt1', 'pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6',
               'limit_bal']

for col in cols_to_cap:
    p1 = df[col].quantile(0.01)
    p99 = df[col].quantile(0.99)
    
    # Create a NEW column with a '_capped' suffix
    df[f'{col}_capped'] = np.clip(df[col], p1, p99)
    

# --- Verify by checking the first few rows ---
display(df[['limit_bal', 'limit_bal_capped', 'bill_amt1', 'bill_amt1_capped']].head())
```

```python
# --- Set up the figure for plotting ---
# We'll look at two key examples: limit_bal and bill_amt1
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Comparing Distributions Before and After Capping Outliers', fontsize=16)

# --- Plot for 'limit_bal' ---
# Original
sns.boxplot(ax=axes[0, 0], y=df['limit_bal'])
axes[0, 0].set_title('Original limit_bal')
axes[0, 0].set_ylabel('Amount')

# Capped
sns.boxplot(ax=axes[0, 1], y=df['limit_bal_capped'], color='skyblue')
axes[0, 1].set_title('Capped limit_bal')
axes[0, 1].set_ylabel('Amount')


# --- Plot for 'bill_amt1' ---
# Original
sns.boxplot(ax=axes[1, 0], y=df['bill_amt1'])
axes[1, 0].set_title('Original bill_amt1')
axes[1, 0].set_ylabel('Amount')

# Capped
sns.boxplot(ax=axes[1, 1], y=df['bill_amt1_capped'], color='lightgreen')
axes[1, 1].set_title('Capped bill_amt1')
axes[1, 1].set_ylabel('Amount')


plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
