# Explanation of the Credit Card Default Dataset

This dataset contains information on default payments, demographic factors, credit data, history of payment, and bill statements of credit card clients in Taiwan from April 2005 to September 2005.

---
## Column Descriptions
---

There are 25 variables in total:

**ID:** A unique identifier for each client.

**LIMIT_BAL:** (Numeric)
- Amount of the given credit in New Taiwan (NT) dollars. It includes both the individual consumer credit and their family (supplementary) credit.

**SEX:** (Categorical)
- Gender of the client.
- 1 = Male
- 2 = Female

**EDUCATION:** (Categorical)
- Client's education level.
- 1 = Graduate school
- 2 = University
- 3 = High school
- 4 = Others
- 5, 6, 0 = Unknown (often grouped into 'Others' during cleaning)

**MARRIAGE:** (Categorical)
- Marital status of the client.
- 1 = Married
- 2 = Single
- 3 = Others
- 0 = Unknown (often grouped into 'Others' during cleaning)

**AGE:** (Numeric)
- Age of the client in years.

---
**PAYMENT HISTORY (April 2005 - September 2005)**
---

**PAY_0, PAY_2, ..., PAY_6:** (Categorical)
- History of past monthly payments. PAY_0 is the repayment status in September 2005, PAY_2 is in August 2005, and so on until PAY_6 for April 2005.
- The measurement scale for the repayment status is:
- -2 = No consumption
- -1 = Paid in full (paid duly)
- 0 = Use of revolving credit
- 1 = Payment delay for one month
- 2 = Payment delay for two months
- ...
- 8 = Payment delay for eight months
- 9 = Payment delay for nine months and above

---
**BILL STATEMENT AMOUNTS (April 2005 - September 2005)**
---

**BILL_AMT1, BILL_AMT2, ..., BILL_AMT6:** (Numeric)
- Amount of bill statement for each month in NT dollars.
- BILL_AMT1 = Amount of bill statement in September, 2005
- BILL_AMT2 = Amount of bill statement in August, 2005
- ...
- BILL_AMT6 = Amount of bill statement in April, 2005

---
**PREVIOUS PAYMENT AMOUNTS (April 2005 - September 2005)**
---

**PAY_AMT1, PAY_AMT2, ..., PAY_AMT6:** (Numeric)
- Amount of the previous payment for each month in NT dollars.
- PAY_AMT1 = Amount paid in September, 2005
- PAY_AMT2 = Amount paid in August, 2005
- ...
- PAY_AMT6 = Amount paid in April, 2005

---
**TARGET VARIABLE**
---

**default payment next month:** (Categorical - Target)
- This is the variable we want to predict.
- 1 = Yes, the client will default on their payment in the next month (October 2005).
- 0 = No, the client will not default.
